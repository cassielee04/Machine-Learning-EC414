**Problem 1:  K-Means**

Here you will apply K-Means on simple Gaussian data. You are given some code to guide you in implementing this clustering method.  Use of online code for K-Means is prohibited; use of the built-in scikitlearn function for K-Means is also prohibited. 
    

**(a)** Generate $3$ $2$D Gaussian clusters of data with the following means and covariances:

$\boldsymbol{\mu}_1 = [2,2]^T, \boldsymbol{\mu}_2 = [-2,2]^T, \boldsymbol{\mu}_3 = [0,-2]^T$,

$\Sigma_1 = [[0.02,0];[0,0.02]]$, $\Sigma_2 = [[0.2,0];[0,0.2]]$, $\Sigma_3 = [[0.05,0];[0,0.05]]$ 

Let each cluster of data have $50$ points. Plot the generated Gaussian data.

```python
import numpy as np
import matplotlib.pyplot as plt
import math as ma

# Generate 2D Gaussian Data with 3 clusters of 3 different means
# Write code here:


DATA = np.zeros((150, 2))

for i in range(0,50):
    DATA[i, 0] = np.random.normal(2, ma.sqrt(.02))
    DATA[i, 1] = np.random.normal(2, ma.sqrt(.02))
for i in range(50,100):
    DATA[i, 0] = np.random.normal(-2, ma.sqrt(.2))
    DATA[i, 1] = np.random.normal(2, ma.sqrt(.2))
for i in range(100,150):
    DATA[i, 0] = np.random.normal(0, ma.sqrt(.05))
    DATA[i, 1] = np.random.normal(-2, ma.sqrt(.05))

# Plotting
# Write code here:

plt.scatter(DATA[:,0], DATA[:,1], marker = "x")
plt.ylabel("Y")
plt.xlabel("X")
plt.grid(True)
plt.xlim(-4.5, 4.5)
plt.ylim(-4.5, 4.5)
```

**(b)** Implement K-Means and apply it to the Gaussian data above. The code below contains comments of what needs to be coded up at a specific part of the code.  Please write your code below these comments where necessary. 

Plot the created clusters per iteration in different colors. For convergence of the algorithm (stopping criterion), check whether the derived cluster means become stationary (don't change over iterations).  

```python
## K-Means implementation:

num_clusters = 3

### CODE 1 - Randomly initialize 3 means: ###
## Write code below here:

#Choose 3 unique random integers in the range 0-149
#These will be the indices of the 3 starting points that belong to the 3 clusters
import random
start = random.sample(range(0,150), num_clusters)
#print(start)
cluster_means = np.zeros((num_clusters, 2))
#C1_muX C1_muY
#C2_muX C2_muY
#C3_muX C3_muY

for i in range(num_clusters):
    cluster_means[i, 0] = DATA[start[i], 0]
    cluster_means[i, 1] = DATA[start[i], 1]
    


##############################################

# initializations
labels = np.ones((DATA.shape[0],1))
converged = 0
iteration = 0
convergence_threshold = 2


while (converged==0):
    iteration += 1
    print('Iteration: ',iteration)
    
    ### CODE 2 - Assignment Step - Assign each data observation to the cluster with the nearest mean: ###
    ## Write code below here:
    
    for i in range(len(labels)):
        dist = np.zeros((1, num_clusters))
        for j in range(num_clusters):
            dist[0, j] = ma.sqrt((DATA[i, 0] - cluster_means[j, 0])**2 + (DATA[i, 1] - cluster_means[j, 1])**2)
        labels[i] = np.argmin(dist, axis=1)
    ###################################################################
    
    
    ### CODE 3 - Mean Updating - Update the cluster means ###
    ## Write code below here:

    cluster_sum = np.zeros((num_clusters, 2))
    cluster_count = np.zeros((num_clusters, 2))
    for i in range(len(labels)):
        if labels[i] == 0:
            cluster_sum[0, 0] += DATA[i, 0]
            cluster_sum[0, 1] += DATA[i, 1]
            cluster_count[0, 0] += 1
            cluster_count[0, 1] += 1
        elif labels[i] == 1:
            cluster_sum[1, 0] += DATA[i, 0]
            cluster_sum[1, 1] += DATA[i, 1]
            cluster_count[1, 0] += 1
            cluster_count[1, 1] += 1
        elif labels[i] == 2:
            cluster_sum[2, 0] += DATA[i, 0]
            cluster_sum[2, 1] += DATA[i, 1]
            cluster_count[2, 0] += 1
            cluster_count[2, 1] += 1
        else:
            print("Invalid label")
    cluster_means = cluster_sum / cluster_count
    
    ###########################################################
    
    ### CODE 4 - Check for convergence ###
    ## Write code below here:

    if iteration == 1:
        conv = float("inf")
    conv_prev = conv
    conv_sum = 0
    for i in range(len(labels)):
        if labels[i] == 0:
            conv_sum += (DATA[i, 0] - cluster_means[0, 0])**2 + (DATA[i, 1] - cluster_means[0, 1])**2
        elif labels[i] == 1:
            conv_sum += (DATA[i, 0] - cluster_means[1, 0])**2 + (DATA[i, 1] - cluster_means[1, 1])**2
        elif labels[i] == 2:
            conv_sum += (DATA[i, 0] - cluster_means[2, 0])**2 + (DATA[i, 1] - cluster_means[2, 1])**2
        else:
            print("Invalid label")
    conv = conv_sum
    print("conv ", conv_prev, " ", conv)
    if ( (conv_prev - conv) < convergence_threshold):
        converged=1
    #############################################################
    
    ### CODE 5 - Plot clustering results after each iteration: ###
    ## Write code below here:

    plt.figure()
    for i in range(len(labels)):
        if labels[i] == 0:
            plt.scatter(DATA[i, 0], DATA[i, 1], color="red")
        elif labels[i] == 1:
            plt.scatter(DATA[i, 0], DATA[i, 1], color="blue")
        elif labels[i] == 2:
            plt.scatter(DATA[i, 0], DATA[i, 1], color="green")
    plt.ylabel("Y")
    plt.xlabel("X")
    plt.grid(True)
    plt.title("K-means Clustering Results")
    plt.show()
    ###################################################################
```

Please write your code below these comments where necessary. Use of online code for DP-Means is prohibited; use of any Python built-in function for DP-Means is also prohibited.

```python
## DP-Means Implementation: ##
import math as ma

# Parameter Initializations
LAMBDA = 8
convergence_threshold = 1
num_points = len(MPG)

### DP Means - Initializations for algorithm ###
# cluster count
K = 1

# sets of points that make up clusters
L = []
L.append(range(num_points))

# Class indicators/labels
Z = np.ones((1,num_points), dtype=np.int)

# means
MU = []
MU.append(np.mean(data_matrix,axis=0))
################################################

# Initializations for algorithm:
converged = 0
t = 0
while converged == 0:
    t += 1
    print('Current iteration: ',t)
    
    K_unchanged = True
    
    ## Per Data Point: ##
    for i in np.arange(num_points):
        
        ### CODE 1 - Calculate distance from current point to all currently existing clusters ###
        ## Write code below here:

        distance = np.zeros((1, len(MU)))
        for j in range(len(MU)):
            distance[0, j] = ma.sqrt((data_matrix[i, 0] - MU[j][0])**2 + (data_matrix[i,1] - MU[j][1])**2)

        
        #########################################################################################
        
        
        ### CODE 2 - Look at how the min distance of the cluster distance list compares to LAMBDA ###
        ## Write code below here:

        closest_cluster = np.argmin(distance, axis = 1)
        if ( distance[0, closest_cluster] > LAMBDA):
            # make new cluster solely consisting of the new point
            K += 1
            Z[0, i] = K
            MU.append(np.array([data_matrix[i,0], data_matrix[i,1]]))
            K_unchanged = False
        #else:
            # assign point to the closest cluster
            Z[0, i] = closest_cluster

        #############################################################################################
    
    ### CODE 3 - Form new sets of points (clusters) ###
    ## Write code below here:


    for i in range(num_points):
        distance = np.zeros((1, len(MU)))
        for j in range(len(MU)):
            distance[0, j] = ma.sqrt((data_matrix[i, 0] - MU[j][0])**2 + (data_matrix[i,1] - MU[j][1])**2)
        
        closest_cluster = np.argmin(distance, axis = 1)
        Z[0, i] = int(closest_cluster)
        
    ##################################################################
    
    ### CODE 4 - Recompute means per cluster ###
    ## Write code below here:
    cluster_sum = np.zeros((K, 2))
    cluster_count = np.zeros((K, 2))
    print(cluster_sum.shape)
    for i in range(num_points):
        
        cluster_sum[Z[0,i], 0] += data_matrix[i, 0]
        cluster_sum[Z[0,i], 1] += data_matrix[i, 1]
        cluster_count[Z[0,i], 0] += 1
        cluster_count[Z[0,i], 1] += 1        
        
    if t==1:
        cluster_means_prev=np.zeros((K,2))
    else:
        cluster_means_prev = cluster_means
    cluster_means = cluster_sum / cluster_count
    #print(cluster_means_prev, cluster_means)
    MU = []
    for i in range(K):
        MU.append(cluster_means[i, :])
    
    ####################################################
    
    ### CODE 5 - Plot clustered data per iteration ###
    ## Write code below here:

    colorlist = []
    plt.figure()
    for i in range(K):
        color=(random.uniform(0, 1), random.uniform(0, 1), random.uniform(0, 1))
        colorlist.append(color)
    for i in range(num_points):
        plt.scatter(data_matrix[i, 0], data_matrix[i, 1], color = colorlist[Z[0,i]])
    plt.show()
    ##########################################################
    
    ### CODE 6 - Test for convergence: number of clusters doesn't change and means stay the same ###
    ## Write code below here:
    
    diff = float("inf")
    if(K_unchanged):
        #print(cluster_means_prev - cluster_means)
        diff = np.sum(abs(cluster_means_prev - cluster_means))
        print(diff)
    if (K_unchanged and diff < convergence_threshold):
        converged = 1
    else:
        converged = 0
    ####################################################



```
